# Aggregate Statistics (GroupBy)

```python
df = pd.read_csv('D:\Programowanie\Workplace\Code\Data_Science_brick_by_brick\DataScienceSummerStudy\_pandas\modified.csv')

numeric_cols = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary']
numeric_df = df[numeric_cols + ['Type 1']]

numeric_df['count'] = 1
# grouped_mean = numeric_df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
# grouped_mean = numeric_df.groupby(['Type 1']).sum()
grouped_mean = numeric_df.groupby(['Type 1']).count()['count']
print(grouped_mean)
```

# Working with large amounts of data

```python
original_df = pd.read_csv('D:\Programowanie\Workplace\Code\Data_Science_brick_by_brick\DataScienceSummerStudy\_pandas\modified.csv')
new_df = pd.DataFrame(columns=original_df.columns)

for chunk in pd.read_csv('D:\Programowanie\Workplace\Code\Data_Science_brick_by_brick\DataScienceSummerStudy\_pandas\modified.csv', chunksize=5):
    results = chunk.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])

print(new_df)
```
