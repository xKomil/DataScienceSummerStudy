# Step 2 : Data Preparation
- Dropping irrelevant columns and rows
- Identyfying duplicated columns
- Identyfing missing values

```python
# df.columns 
## omenting the lines that we dont wanna keep
df = df[['coaster_name', 
    # 'Length', 'Speed', 
    'Location', 'Status', 
    # 'Opening date',
       #'Type', 
       'Manufacturer', 
       #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 
       #'Opened', 
       #'Replaced by', 'Website',
    #    'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #    'Single rider line available', 'Restraint Style',
    #    'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
       'latitude', 'longitude', 
       'Type_Main',
       'opening_date_clean',
    #    'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    #    'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
# Or we can do it with .drop() axis=1 = column axis=0 = row
df.drop(['Opening date'], axis=1)
```

```python
df.shape
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
# df['year_introduced] = pd.to_numeric(['year_introduced])
df.dtypes
```

```python
# Rename our columns
df = df.rename(columns={'coaster_name':'Coaster_Name',
          'year_introduced':'Year_Introduced',
          'opening_date_clean':'Opening_Date',
          'height_ft':'Height_ft',
          'latitude':'Latitude',
          'longitude':'Longitude',
          'speed_mph':'Speed_mph',
          'Inversions_clean':'Inversions',
          'Gforce_clean': 'Gforce'})
df.columns
```

```python
# Identyfing missing values for column
df.isna().sum()
```

```python
# Searching for duplicated data
df.loc[df.duplicated()]
```

```python
# df.duplicated(subset=['Coaster_Name'])
df.loc[df.duplicated(subset=['Coaster_Name'])].head()
```

```python
# Watching why the are duplicated names values
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
# df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date']).sum()
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True).copy()
df
```

Step 3 : Feature Understanding
(univariate analysis) - 

- Plotting Feature Distributions
- Histogram
- Kde - estymacja gęstości jądra [kde](https://manuals.pqstat.pl/statpqpl:plotpl:kde1pl)
- Boxplot

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar', title='Top 10 Years Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
# Wyświetlanie wykresu
plt.show()
```

```python
ax = df['Speed_mph'].plot(kind='hist', bins=20, title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```

```python
ax = df['Speed_mph'].plot(kind='kde', title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```

```python
ax = df['Speed_mph'].plot(kind='box', title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
plt.show()
```
