### 3. Visualizing Distribution of the data
- histplot()
- kdeplot()
- jointplot()
- rugplot()

```python
x = random(100)
sns.histplot(x)
sns.displot(x, kde=True)
```

```python
sns.displot(x, kde=True, bins=30)
```

```python
sns.kdeplot(x, fill=True, cbar=True)
```

```python
sns.kdeplot(x, fill=True, cbar=True, bw_adjust=1, cut=15)
```

```python
x = tips['total_bill']
y = tips['tip']
sns.jointplot(x=x, y=y)
```

```python
sns.jointplot(x=x, y=y, kind='hex')
```

```python
sns.jointplot(x=x, y=y, kind='kde',ratio=10)
```

```python
f, ax = plt.subplots(figsize=(6,6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(x=x, y=y, cmap=cmap, n_levels=60, fill=False)
```

```python
g = sns.jointplot(x=x, y=y, kind='kde', color='m', fill=True)
g.plot_joint(plt.scatter, c = 'w', s = 30, linewidth = 1, marker = '+')
g.ax_joint.collections[0]
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
```
